# Linear Regression in Python

## Hypotheses:
### 1. Does curfew type in teen years affect lifetime arrests?
Curfew types: set by parent, set by child, set jointly

* H1: There is a significant correlation between curfew type in teen years and lifetime arrests.
* H0: There is no significant correlation between curfew type in teen years and lifetime arrests.

### 2. Does parenting style experienced in teen years affect lifetime arrests?
Parenting styles: authoritative, authoritarian, permissive

* H1: There is a significant correlation between the parenting style experienced in teen years and lifetime arrests.
* H0: There is no significant correlation between the parenting style experienced in teen years and lifetime arrests.
-----

## The data:
* Population: U.S. youth ages 13 - 17 in 1997 that live with a parental figure
* Sample: data from National Longitudinal Study - Youth 1997 (NLSY97) of the Bureau of Labor Statistics (BLS)

* Data is oversampled for blacks
   * Oversampling will make regression results for blacks--a smaller group--more accurate than without oversampling
   * However, descriptive statistics won't accurately reflect the population without weights to correct for oversampling
   
-----
# Types: A (authoritarian), H (authoritative, healthy), P (permissive), U (uninvolved)

## Exploratory data analysis for df_curfew

The age-old question: how much does parenting style affect children once they become adults?

### Descriptive statistics
We have n = 2600

#### Criminal variables
The criminal variables total arrests (ARREST_TOT), total incarcerations (INCAR_TOT), and total months incarcerated (INCAR_TOT_MO) are cumulative over the lifetime of the respondents who were all born in the mid-80s. One or more of these will serve as the dependent variable (in one or more regressions). The independent variables are all data collected in 1997 when the repondents were 13 or younger.

From the descriptive statistics of our criminal variables, we can see total arrests (ARREST_TOT) and total months incarcerated (INCAR_TOT_MO) have relatively high maximums compared to total incarcerations (INCAR_TOT), which are 62, 192, and 7 respectively. This leaves little room for variance in the INCAR_TOT variable. Using total months incarcerated as a kind of proxy for the severity of incarceration will likely make it a better variable than INCAR_TOT, and more comparable to ARREST_TOT.

Given this information, I plan on using both ARREST_TOT and INCAR_TOT_MO as the dependent variable in separate regressions to see if parenting curfew style is correlated with either.

#### Gang-related variables
I wanted to account for youth already exposed to or involved in criminal activity at a young age for more accurate parenting curfew style estimators. To do this, I am going to include ARREST_TOT97 which gives the total number of arrests in 1997 when all respondents were 13 or younger. I'm also going to include a gang-related variable, though I will need to test for multicollinearity.

For our gang-related variables, the mean in highest for a gang in the neighborhood or school at 41% (NEIGH_SCH_GANG) followed by 38% with a sibling or friend in a gang (SIB_FR_GANG), almost 20% for percent peers in a gang (PERC_PEERS_GANG), and almost 19% of respondents having ever been in a gang (EV_GANG).

Since EV_GANG would likely correlate most highly with criminal activity, I will likely use this variable in the model.

#### Economic variables
Since much crime is economically motivated, I wanted to account for the respondents' economic situation for more accurate parenting style estimators. 

To do this, I have HARD_TIMES97, which asks the respondent if they've ever been homeless, and HH_INCOME97, the household income for the respondent in 1997. Like with the EV_GANG and ARREST_TOT97, I'll need to watch for multicollinearity. 

For HH_INCOME97, I may need to log it to make it linear for the regression.

#### Demographic variables
Using sex and race variables, I created six interacted dummy variables: black males, black females, Hispanic males, Hispanic females, white males, and white females.

I know from prior criminal justice research how important it is to interact race, sex, and age to fully and more accurately capture the effects for these groups. We can see this below in the correlation matrix for each group's correlation with the criminal variables. Thankfully, the cohort is all within a few years of each other in age.

I will drop one of these dummy variables for the regression to create a reference group.

#### Parenting style variables
Parenting styles include the child setting their own curfew (C), a joint parent-child set curfew (J), and a parent setting the curfew (P). Like with the race/sex variables, I'll need to drop one of these in the regression to avoid the dummy variables trap.

It's important to note that these groups are drastically different sizes. Children setting their own curfew is only 2.6% of the respondents. Joint-set curfew is about 31% and parents setting the curfew makes up the rest at about 47%. Because our sample size of the P group is so large, the coefficients will likely be more trustworthy than the small sample size of the C group.

* Unsurprisingly, black males have the highest positive correlation with ARREST_TOT, INCAR_TOT, and INCAR_TOT_MO out of the six sex/race groups. On the flipside, white females have the greatest negative correlation with those variables.

```python
def label_parenting(row):
   if row['U'] == 1 :
      return 'U'
   if row['P'] == 1 :
      return 'P'
   if row['N'] == 1 :
      return 'N'
   if row['A'] == 1:
      return 'A'
   else:
       return 0

df_curfew['PARENTING'] = df_curfew.apply(lambda row: label_parenting(row), axis=1)

print(df_curfew.columns)
df_curfew['U'].value_counts()
```
